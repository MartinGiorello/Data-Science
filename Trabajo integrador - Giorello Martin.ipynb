```python
#creamos funcion para realizar grafico de barras con pandas que nos va a servir para features con valores categóricos
def bar_counts (columna):
    """la función crea un diagrama de barra capturando excepciones
    """
    try:  
        counts = stroke[columna].value_counts()
       
        counts.plot.bar (color="green")
        
        plt.title (columna)
       
        plt.ylabel ("Cantidad")
    except TypeError:
        print('Revisar el tipo de dato ingresado')
    except SyntaxError:
        print('Revisar la sintaxis')
    except Exception as error:
        print(f"No se pudo realizar el gráfico por el siguiente error: {error}")
```

### *Columnas calculadas*

```python
#creo la funcion y su documentación
def agrupar_bmi (bmi:int):
    """esta función toma los valores de la columna bmi y los cataloga como bajo, medio o alto y crea un nuevo feature categórico ordinal con el fin de agrupar los valores en 3 categorias
    """
    try:
        if bmi < 0:
            return "Inválido"
        elif 0 < bmi < 20:
            return "Bajo"
        elif 20 < bmi < 40:
            return "Medio"
        elif bmi >= 40:
            return "Alto"
    except TypeError:
        print('Revisar el tipo de dato ingresado')
    except SyntaxError:
        print('Revisar la sintaxis')
    except Exception:
        print(f'No se pudo realizar el gráfico por el siguiente error: {Exception}')


#creo la funcion y su documentación
def agrupar_glucose (avg_glucose_level:int):

    """esta función toma los valores de la columna avg_glucose_level y los cataloga como bajo, medio o alto y crea un nuevo feature categórico ordinal  con el fin de agrupar los valores en 3 categorias
    """
    try:
        if avg_glucose_level < 0:
            return "Inválido"
        elif 0 < avg_glucose_level < 100:
            return "Bajo"
        elif 100 < avg_glucose_level < 200:
            return "Medio"
        elif avg_glucose_level >= 200:
            return "Alto"
    except TypeError:
        print('Revisar el tipo de dato ingresado')
    except SyntaxError:
        print('Revisar la sintaxis')
    except Exception:
        print(f'No se pudo realizar el gráfico por el siguiente error: {Exception}')
```

```python
#creo dos features nuevos en el dataset llamados bmi_agrupado y AVG_agrupado que van a ser el resultado de la implementacion de las funciones previamente creadas
stroke ["bmi_agrupado"] = stroke ["bmi"].apply(agrupar_bmi)
stroke ["AVG_agrupado"] = stroke ["avg_glucose_level"].apply(agrupar_glucose)
```

```python
#podemos practicar un scatterplot que nos visualice la dispersion de los datos en el feature de la muestra, resaltando en otro color los outliers
#creamos funcion que nos va a servir tanto para la columna "bmi" como para "avg_glucose_level"
def dispersion_outliers (ds, columna, titulo):
  
    #calculo de cuantiles y umbrales
    Q1 = ds[columna].quantile(0.25)
    
    Q3 = ds[columna].quantile(0.75)
    
    IQR = Q3 - Q1
    
    umbralSup = Q3 + 1.5 * IQR
    
    umbralInf = Q1 - 1.5 * IQR
    #defino que todos los outliers serán o bien menores al umbral inferior o bien mayores al umbral superior.
    outliers = ds[(ds[columna] > umbralSup) | (ds[columna] < umbralInf)]
  
    plt.scatter(ds.index, ds[columna], c="skyblue", alpha=0.5, label='Index')

    plt.title (titulo)
    
    plt.xlabel("Indice")
        
    plt.ylabel(columna)

    plt.suptitle ("Stroke Dataset")
    
    plt.grid(axis='y', linestyle='--', alpha=0.8)
    
    plt.scatter(outliers.index,outliers[columna], c="red", label='Outliers')
    
    plt.legend()
    
    plt.show()
```

```python
dispersion_outliers (stroke, "bmi", "Indice masa corporal")
```

```python
dispersion_outliers (stroke, "avg_glucose_level", "Nivel promedio de glucosa")
```
